### 6. Multi-Answer Survey Questions


- Analyzing survey data may also involve dealing with questions that have mutliple answers. 
- In this survey, participants were questioned regarding the programming languages they used and advised to check as many choices as they felt were applicable.

6.1 Let's try to figure out how many times each language was mentioned, find developers' favorite, and which ones are less popular. 

```python
df["LanguageWorkedWith"].head(4)
```

- The result above clearly shows that the responses were stored in a single column with *;* as the delimiter.
- It will be necessary to perform string splitting to put each language in its own column in a new dataframe for languages.

```python
langs_list = df["LanguageWorkedWith"].str.split(";", expand=True)
```

```python
langs_list.head()
```

- By doing this, we have assigned each language its own column, yet each row still represents one respondent. 

6.2 To find the number of times a language was mentioned, we cannot use *value_counts()* in this case, since it only works on panda Series:
- To this end, we will first need to stack this Dataframe by slicing each column and stacking them on top of each other so that every data point appears in a single panda Series.
- Finally, we will now be able to use *value_counts()* on the newly stacked series to get the total count each language was mentioned. 

```python
pop_langs = langs_list.stack().value_counts()
```

```python
pop_langs
```

- From the results obtained above, JavaScript appears to have been the most popular language among developers in 2019 at the time of the survey. 

6.3 Let's plot the results above to get a better visual representation of the results. 

```python
pop_langs.plot(kind='bar', figsize=(20, 8))
```
